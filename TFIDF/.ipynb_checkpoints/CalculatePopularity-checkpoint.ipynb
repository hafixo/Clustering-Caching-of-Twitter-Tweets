```python
tweets["Cleaned"] = tweets["Cleaned_Initial"]
```

```python
sentence_to_list = lambda x:x.split(" ")
tweets['sentences_split'] = tweets["Cleaned_Initial"].apply(sentence_to_list)
```

```python
index = 0
def Calculate_Popularity_single(text):
    global index
#     text = text.Cleaned_Initial
    totalvalue = 0
    count = 0
    for singleword in text:
        try:
            value = vocab_idf[vocab_idf["Unnamed: 0"] == singleword]["tfidf"].values[0]
            totalvalue = totalvalue+value
            count = count+1
        except:
            pass
    if(count > 0):
        calcualted_popularity = totalvalue/count
    else:
        calcualted_popularity = -1
    index = index+1
    if(index % 10 == 0):
        print("Processed: ", index)
    return calcualted_popularity
```

```python
popularity = lambda x:Calculate_Popularity_single(x)
```

```python
#Parallel Execution
# with mp.Pool(mp.cpu_count()) as pool:
#     df['sentences_score'] = pool.map(Calculate_Popularity_single, tweets['sentences_split'])
```

```python
tweets['sentences_score'] = tweets["sentences_split"].apply(Calculate_Popularity_single)
```

```python
def Calculate_Popularity():
    Popularity = []
    sample_text = []
    for singletext in text:
        splittedtext = singletext.split(" ")
        totalvalue = 0
        count = 0
        for singleword in splittedtext:
            try:
                value = vocab_idf[vocab_idf["Unnamed: 0"] == singleword]["tfidf"].values[0]
                totalvalue = totalvalue+value
                count = count+1
            except:
                pass
        if(count > 0):
            calcualted_popularity = totalvalue/count
            Popularity.append(calcualted_popularity)
            sample_text.append(splittedtext)
        else:
            Popularity.append(-1)
            sample_text.append(splittedtext)
    return Popularity,sample_text
```
